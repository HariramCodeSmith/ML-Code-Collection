```python
import networkx as nx
from networkx.algorithms import bipartite
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
# you need to have tensorflow 
from stellargraph.data import UniformRandomMetaPathWalk
from stellargraph import StellarGraph
```

```python
data=pd.read_csv('movie_actor_network.csv', index_col=False, names=['movie','actor'])
```

```python
edges = [tuple(x) for x in data.values.tolist()]
```

```python
movie_dict = dict()
for i in edges:
    if(i[0] in movie_dict):
        movie_dict[i[0]].append(i[1])
    else:   
        movie_dict[i[0]] = [i[1]]
```

```python
B = nx.Graph()
B.add_nodes_from(data['movie'].unique(), bipartite=0, label='movie')
B.add_nodes_from(data['actor'].unique(), bipartite=1, label='actor')
B.add_edges_from(edges, label='acted')
```

```python
B
```

```python
A = list(nx.connected_component_subgraphs(B))[0]
```

```python
A = (B.subgraph(c) for c in nx.connected_components(B))
A = list(A)[0]
```

```python
print("number of nodes", A.number_of_nodes())
print("number of edges", A.number_of_edges())
```

```python
movies = []
actors = []
for i in A.nodes():
    if 'm' in i:
        movies.append(i)
    if 'a' in i:
        actors.append(i)
print('number of movies ', len(movies))
print('number of actors ', len(actors))
```

```python

# Create the random walker
rw = UniformRandomMetaPathWalk(StellarGraph(A))

# specify the metapath schemas as a list of lists of node types.
metapaths = [
    ["movie", "actor", "movie"],
    ["actor", "movie", "actor"]
]

walks = rw.run(nodes=list(A.nodes()), # root nodes
               length=100,  # maximum length of a random walk
               n=1,        # number of random walks per root node 
               metapaths=metapaths
              )

print("Number of random walks: {}".format(len(walks)))
```

```python
from gensim.models import Word2Vec
model = Word2Vec(walks, vector_size=128, window=5)
```

```python
model.wv.vectors.shape  # 128-dimensional vector for each node in the graph
```

```python
# Retrieve node embeddings and corresponding subjects
node_ids = model.wv.index_to_key  # list of node IDs
node_embeddings = model.wv.vectors  # numpy.ndarray of size number of nodes times embeddings dimensionality
node_targets = [ A.nodes[node_id]['label'] for node_id in node_ids]
```

image.png

```python
def data_split(node_ids,node_targets,node_embeddings):
    '''In this function, we will split the node embeddings into actor_embeddings , movie_embeddings '''
    actor_nodes,movie_nodes=[],[]
    actor_embeddings,movie_embeddings=[],[]
    for i,j,k in zip(node_embeddings,node_targets,node_ids):
        if(j == 'actor'):
            actor_embeddings.append(i)
            actor_nodes.append(k)
        else:
            movie_embeddings.append(i)
            movie_nodes.append(k)
            
    # split the node_embeddings into actor_embeddings,movie_embeddings based on node_ids
    # By using node_embedding and node_targets, we can extract actor_embedding and movie embedding
    # By using node_ids and node_targets, we can extract actor_nodes and movie nodes

    return actor_nodes,movie_nodes,actor_embeddings,movie_embeddings
```

```python
import matplotlib.pyplot as plt

def cost1(graph,number_of_clusters):
    '''In this function, we will calculate cost1'''
    actor_dict = dict()
    for k in list(graph.edges()):
        if(k[1] in actor_dict):
            actor_dict[k[1]].append(k[0])
        else:   
            actor_dict[k[1]] = [k[0]]
    
    embed = list()
    li = list()
    for j in graph.nodes():
        if 'a' in j:
            li.append(actor_nodes.index(j))
            embed.append(actor_embeddings[actor_nodes.index(j)])
    #plt.figure(figsize=(20,5))
    
    kmeans = KMeans(n_clusters=number_of_clusters, max_iter=30,random_state=0).fit(np.array(embed))
        
    labels = kmeans.labels_
    
    df = pd.DataFrame()
    df['actor_node_index'] = li
    df['labels'] = labels
    sum_l = 0
    for h in range(number_of_clusters):
        index_values = df[df['labels'] == h]['actor_node_index'].values
        g = nx.Graph()
        edge_li = list()
        movies_ver = set()
        for v in index_values:
            val_list = actor_dict[actor_nodes[v]]
            for m in val_list:
                movies_ver.add(m)
                c = list()
                c.append(actor_nodes[v])
                c.append(m)
                edge_li.append(tuple(c))
       
        g.add_nodes_from(np.array(actor_nodes)[index_values], bipartite=0, label='actor')
        g.add_nodes_from(movies_ver, bipartite=1, label='movie')
        g.add_edges_from(edge_li, label='acted')
        #l= set(np.array(actor_nodes)[index_values])
        #r= movies_ver
        #pos = {}
        #pos.update((node, (1, index)) for index, node in enumerate(l))
        #pos.update((node, (2, index)) for index, node in enumerate(r))
        #nx.draw_networkx(g, pos=pos, with_labels=True,node_color='lightblue',alpha=0.8,style='dotted',node_size=500)
        #plt.show()
        tot_nodes = len(g.nodes())
        edge = 0
        count = 0
        subgr = list(g.subgraph(c) for c in nx.connected_components(g))
        
        for y in subgr:
            if(edge<y.number_of_edges()):
                edge = y.number_of_edges()
                edge_ind = count 
            count+=1
        node_count = len(list(subgr[edge_ind].nodes()))
        sum_l = sum_l + (node_count/tot_nodes)
        #print(node_count,tot_nodes)
    cost1 = sum_l * (1/number_of_clusters)
    return cost1
```

```python
import matplotlib.pyplot as plt

def cost2(graph,number_of_clusters):
    '''In this function, we will calculate cost2'''
    movie_dict = dict()
    for k in list(graph.edges()):
        if(k[0] in movie_dict):
            movie_dict[k[0]].append(k[1])
        else:   
            movie_dict[k[0]] = [k[1]]
    
    embed = list()
    li = list()
    for j in graph.nodes():
        if 'm' in j:
            li.append(movie_nodes.index(j))
            embed.append(movie_embeddings[movie_nodes.index(j)])
    
    
    kmeans = KMeans(n_clusters=number_of_clusters, max_iter=30,random_state=0).fit(np.array(embed))
        
    labels = kmeans.labels_
    
    df = pd.DataFrame()
    df['movie_node_index'] = li
    df['labels'] = labels
    sum_l = 0
    for h in range(number_of_clusters):
        index_values = df[df['labels'] == h]['movie_node_index'].values
        g = nx.Graph()
        edge_li = list()
        actor_ver = set()
        for v in index_values:
            val_list = movie_dict[movie_nodes[v]]
            for a in val_list:
                actor_ver.add(a)
                c = list()
                c.append(movie_nodes[v])
                c.append(a)
                edge_li.append(tuple(c))
       
        g.add_nodes_from(np.array(movie_nodes)[index_values], bipartite=0, label='actor')
        g.add_nodes_from(actor_ver, bipartite=1, label='movie')
        g.add_edges_from(edge_li, label='acted')
        
        tot_nodes = len(g.nodes())
        edge = 0
        count = 0
        subgr = list(g.subgraph(c) for c in nx.connected_components(g))
        
        for y in subgr:
            if(edge<y.number_of_edges()):
                edge = y.number_of_edges()
                edge_ind = count 
            count+=1
        node_count = len(list(subgr[edge_ind].nodes()))
        sum_l = sum_l + (node_count/tot_nodes)
        
    cost2 = sum_l * (1/number_of_clusters)
    return cost2
```
